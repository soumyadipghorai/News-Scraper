```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np 
import random
```

```python
indiaNewsUrl = "https://www.indiatoday.in/india"
indiaNewspage = requests.get(indiaNewsUrl)
indiaNewsHtmlContent = indiaNewspage.content
indiaNewsSoup = BeautifulSoup(indiaNewsHtmlContent, "html.parser")

indiaNewsDiv = indiaNewsSoup.find_all("div", {"class" : "catagory-listing"})

for div in indiaNewsDiv : 

    imageDiv = div.find("div", {"class" : "pic"})
    image = div.find('img')
    print(image.get('src'))

    header = div.find("h2")
    print(header.text)

    para = div.find("p")
    print(para.text)

    link = div.find("a")['href']
    print(link)
```
